# Data Visualization using python


Python provides <code>matplotlib</code> and <code>seaborn</code> libraries to visualize data. 
<br>
<code>matplotlib</code> provides high quality 2D and 3D basic charts like line charts, barcharts etc.
<br>
<code>seaborn</code>, built on top of matplotlib provides more complex charts and aesthetics.

### Import required libraries

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

<br>

### Load the iris dataset

The Iris dataset consists of 150 entries from 3 species of iris flowers. It contains the following dimensions:
- sepal_length
- sepal_width
- petal_length
- petal_width
- species

```python
iris=sns.load_dataset('iris')
iris
```

<br>

### General statistical summary of the Iris data:

#### Using pandas.describe()

```python
iris.describe()
```

<br>

#### using seaborn.pairplot()

```python
sns.pairplot(iris,hue='species',height=3.5,palette=['mediumpurple','c','rosybrown'])
plt.show()
```

<br>

### Pie Plot for Species Frequency:

```python
# extract count of each species in the dataset
count=iris['species'].value_counts()
count
```

```python
# construct the pie plot
plt.figure(figsize=(4,4))
plt.pie(count,labels=count.index,colors=['mediumpurple','c','rosybrown'] ,autopct='%.3f%%',explode=(0.02,0.02,0.02),startangle=90)
plt.title('Frequency of Species',color='indigo')
plt.show()
```

<br>

### Relationship Between Sepal Length and Width:

```python
plt.figure(figsize=(9,6))
sns.scatterplot(x='sepal_length',y='sepal_width',hue='species',palette=['mediumpurple','c','rosybrown'],data=iris,marker='.',s=100)
plt.title('Sepal length vs Sepal width',color='indigo')
plt.xlabel('sepal length',color='indigo')
plt.ylabel('sepal width',color='indigo')
plt.show()
```

<br>

### Distribution of Sepal and Petal Features:

```python
sns.histplot(data=iris,x='sepal_length',binwidth=0.15,color='lightsteelblue')
plt.title('Distribution of sepal length',color='indigo') 
plt.ylabel('count',color='indigo')
plt.xlabel('sepal length',color='indigo')
plt.show()
```

```python
sns.histplot(data=iris,x='sepal_width',binwidth=0.15,color='lightsteelblue')
plt.title('Distribution of sepal width',color='indigo') 
plt.ylabel('count',color='indigo')
plt.xlabel('sepal width',color='indigo')
plt.show()
```

```python
sns.histplot(data=iris,x='petal_length',binwidth=0.25,color='lightsteelblue')
plt.title('Distribution of petal length',color='indigo') 
plt.ylabel('count',color='indigo')
plt.xlabel('petal length',color='indigo')
plt.show()
```

```python
sns.histplot(data=iris,x='petal_width',binwidth=0.15,color='lightsteelblue')
plt.title('Distribution of petal width',color='indigo') 
plt.ylabel('count',color='indigo')
plt.xlabel('petal width',color='indigo')
plt.show()
```

<br>

### Jointplot of Sepal Length vs Sepal Width:

```python
plot=sns.jointplot(x='sepal_length',y='sepal_width',data=iris,kind='scatter',hue='species',palette=['mediumpurple','c','rosybrown'],height=6)
plot.fig.suptitle('Joint plot between Sepal length and sepal width',color='indigo',y=1.02)
plt.xlabel('sepal length',color='indigo')
plt.ylabel('sepal width',color='indigo')
plt.show()
```

<br>

### KDE Plot for Setosa Species (Sepal Length vs Sepal Width):

```python
setosa=iris[iris['species']=='setosa']
sns.kdeplot(x='sepal_length',y='sepal_width',data=setosa,fill=True,color='mediumpurple')
plt.title('KDE plot between sepal length and sepal width of setosa species',color='indigo')
plt.xlabel('sepal length',color='indigo')
plt.ylabel('sepal width',color='indigo')
plt.show()
```

<br>

### KDE Plot for Setosa Species (Petal Length vs Petal Width):

```python

sns.kdeplot(x='petal_length',y='petal_width',data=setosa,fill=True,color='mediumpurple')
plt.title('KDE plot between petal length and petal width of setosa species',color='indigo')
plt.xlabel('petal length',color='indigo')
plt.ylabel('petal width',color='indigo')
plt.show()
```
